```python
#removing column_2 because it has no variance  
data1.drop(columns=['feature2'], inplace=True)
```

```python
data1.head(3)
```

```python
#Chnaging the Label 
encoding = {'g': 1, 'b': 0}
data1.label.replace(encoding, inplace=True)
data1
```

```python
data1.info()
```

```python
#Changing the feature1 into numeric 
data1['feature1'] = data1.feature1.astype('float64')
```

```python
data1.info()
```

```python
#training and testing the Data 
X = data1.values[:, :-1]
y = data1.values[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,  train_size=0.50,test_size=0.30,random_state=42)
```

```python
from keras import models
from keras import layers
def model_build():
    
  model = models.Sequential()
  model.add(layers.Dense(16, input_dim=33, activation='relu'))     # kernel_initializer='normal'
  model.add(layers.Dense(8, activation='relu'))                 # kernel_initializer='normal'
  model.add(layers.Dense(1, activation='relu'))                 # kernel_initializer='normal'
  return model
model=model_build()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=1)
test_loss, test_acc = model.evaluate(X_test, y_test)
```

```python
print('Test accuracy:', test_acc)
```

```python
y_predict=model.predict(X_test)
y_predict
```
